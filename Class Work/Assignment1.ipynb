```python
# Taking two numbers input from user
num1 = int(input("Enter the first number :"))
num2 = int(input("Enter the second number :"))

#Check if both numbers are even or odd or one of each
if num1 % 2 == 0 and num2 % 2 == 0:
    print("Both numbers are even numbers.")

elif num1 % 2 != 0 and num2 % 2 != 0:
    print("Both numbers are odd numbers.")

elif num1 % 2 == 0 and num2 % 2 != 0 :
    print("First number is even but second number is odd.")

elif num1 % 2 != 0 and num2 % 2 == 0 :
    print("First number is odd but second number is even.")

```
